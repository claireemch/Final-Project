# Tree Data Set

If you have ever taken a dog on a walk, then you know that they love smelling all kinds of different trees.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

info = pd.read_csv("trees.csv" , index_col = "id", low_memory=False)
df = pd.DataFrame(info)
set = df.groupby('neighborhood').size().sort_values(ascending = False)
set.head(20)
```

```python
set.rank(ascending = False)
```

The csv file gives a list of trees without telling you how many trees each neighborhood has. Therefore, I had to group the trees by neighborhoods to see how many trees each neighborhood has.

# But

grouping the trees by neighborhoods wasn't enough because some trees are in a resedential, commercial, or industrial properties. Accordingly, I only take a look at the different kinds of trees in parks based on the neighborhood since we don't want to trespass.

```python
park_mask = df["land_use"]== "Park"
parks = df[park_mask]

parks["neighborhood"].value_counts().head(10)
```

```python
parks["neighborhood"].value_counts().head(15).plot(kind='bar', figsize=(15,5), color='g')
```

```python
parks["neighborhood"].value_counts().head().plot(kind='pie', figsize=(6,6) , autopct = "%.2f")
```
